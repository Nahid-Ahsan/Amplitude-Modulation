Consider, Information signal (fm) = 10 Hz and Carrier signal (fc ) = 100 Hz.

```python
import numpy as np
from numpy import cos, absolute
from numpy.fft import fft,fftshift
from math import pi
import matplotlib.pyplot as plt
```

                                        DSB_FC

```python
fs = 8000 # Sampling Frequency
fc = 100 # Carrier Frequecy
Ac = 1 # Carrier Amplitude
fm = 10 # Message Frequency
Am = 0.6 # Message Amplitude
```

```python
t = np.arange(0,1,1/fs) # Time Grid
M = Am*cos(2*pi*fm*t) # Message Signal
C = Ac*cos(2*pi*fc*t) # Carrier Signal
S_AM = (Ac+M)*cos(2*pi*fc*t) # DSB+C Signal
```

```python
plt.plot(t,M)
plt.xlabel('Time (s)')
plt.ylabel('m(t)')
plt.xlim((0,5/fm))
plt.title('Message Signal in Time Domain')
plt.show()
```

```python
plt.plot(t,C)
plt.xlabel('Time (s)')
plt.ylabel('c(t)')
plt.xlim((0,5/fm))
plt.title('Carrier Signal in Time Domain')
plt.show()
```

```python
plt.plot(t,S_AM)
plt.xlabel('Time (s)')
plt.ylabel('S_{AM}(t)')
plt.xlim((0,5/fm))
plt.title('DSB+C Signal in Time Domain')
plt.show()
```

```python
M_f = fftshift(absolute(fft(M)))
f = np.linspace(-fs/2, fs/2, len(M_f))
```

```python
plt.stem(f, M_f/M_f.max())
plt.xlabel('Frequency (Hz)')
plt.ylabel('Normalized |M(f)|')
plt.xlim((-120,120))
plt.title('Spectrum of Message Signal')
plt.show()
```

```python
C_f = fftshift(absolute(fft(C)))
```

```python
plt.stem(f, C_f/C_f.max())
plt.xlabel('Frequency (Hz)')
plt.ylabel('Normalized |C(f)|')
plt.xlim((-120,120))
plt.title('Spectrum of Carrier Signal')
plt.show()
```

```python
S_AM_f = fftshift(absolute(fft(S_AM)))
```

```python
plt.stem(f, S_AM_f/S_AM_f.max())
plt.xlabel('Frequency (Hz)')
plt.ylabel('Normalized |S_{AM}(f)|')
plt.xlim((-120,120))
plt.title('Spectrum of DSB+C Signal')
plt.show()
```
